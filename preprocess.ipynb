Calculate occupancy duration per bin: For each charging event, calculate the duration it occupied within each time bin it overlaps with.

```python
# --- 2. 점유율 계산을 위한 함수 정의 ---
def calculate_occupancy(df_row):
    """
    하나의 충전 기록(행)을 받아, 각 1시간 단위 슬롯별 점유 시간을 계산합니다.
    """
    start = pd.to_datetime(df_row['startedAt'])
    end = pd.to_datetime(df_row['endedAt'])

    # 충전이 시작된 시간의 '정시'를 기준으로 첫 번째 슬롯을 잡음
    # 예: 08:40 -> 08:00
    current_slot_start = start.floor('H')

    occupancy_records = []

    # 충전이 끝날 때까지 1시간 단위로 루프를 돔
    while current_slot_start < end:
        current_slot_end = current_slot_start + pd.Timedelta(hours=1)

        # 현재 슬롯 내에서 실제 점유 시작/종료 시간 계산
        # max(슬롯 시작, 실제 충전 시작), min(슬롯 끝, 실제 충전 끝)
        overlap_start = max(current_slot_start, start)
        overlap_end = min(current_slot_end, end)

        # 겹치는 시간(분) 계산
        overlap_minutes = (overlap_end - overlap_start).total_seconds() / 60.0


        if overlap_minutes > 0:
            occupancy_records.append({
                'statId': df_row['alternateStatId'],
                'charger':df_row['_charger'],
                'time_slot': current_slot_start,
                'occupied_minutes': overlap_minutes
            })

        # 다음 시간 슬롯으로 이동
        current_slot_start += pd.Timedelta(hours=1)

    return occupancy_records
```

```python
all_occupancies_list = df.apply(calculate_occupancy, axis=1).tolist()
```

```python
# 2차원 리스트를 1차원으로 펼침
flat_list = [item for sublist in all_occupancies_list for item in sublist]

# 최종 점유 기록 데이터프레임 생성
occupancy_df = pd.DataFrame(flat_list)
```

```python
chargers_per_slot = occupancy_df.groupby(['statId', 'time_slot'])['charger'].count().reset_index(name='num_of_chargers')
chargers_per_slot
```

```python
demand = occupancy_df.groupby(['statId', 'time_slot'])['charger'].nunique().reset_index(name='demand')
```

```python
# Calculate the number of unique chargers per station and time slot
aggregated_occupancy = occupancy_df.groupby(['statId', 'time_slot'])['occupied_minutes'].sum().reset_index()
# 같은 statId, time_slot이 한 개만 있는지 확인
```

```python
aggregated_occupancy.head()
```

```python
# Merge the number of chargers into the aggregated occupancy dataframe
aggregated_occupancy = pd.merge(aggregated_occupancy, chargers_per_slot, on=['statId', 'time_slot'], how='left')

print(aggregated_occupancy.head())
```

```python
aggregated_occupancy = pd.merge(aggregated_occupancy, demand, on=['statId', 'time_slot'], how='left')
```

```python
aggregated_occupancy
```

```python
aggregated_occupancy['occupied_rate'] = aggregated_occupancy['occupied_minutes'] / (aggregated_occupancy['num_of_chargers'] * 60)
```

```python
final_pivot = aggregated_occupancy.pivot(index='time_slot', columns='statId', values=['occupied_rate', 'demand'])

# 전체 시간 범위를 생성하여 비어있는 시간(NaN)은 0으로 채움
# 데이터의 시작과 끝 시간의 정시를 기준으로 전체 시간 범위를 만듦
full_time_range = pd.date_range(
    start=pd.to_datetime(df['startedAt'].min()).floor('H'),
    end=pd.to_datetime(df['endedAt'].max()).ceil('H'),
    freq='H'
)
final_pivot = final_pivot.reindex(full_time_range).fillna(0)
```
